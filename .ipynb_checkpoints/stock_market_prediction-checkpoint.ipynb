# Predicting the S&P 500

The goal of this project is to accurately predict the price of the S&P 500 Index based off of historical performance. The data I've used includes the daily opening price and closing price of the S&P500 from 1950 to 2015. I will train the model using data from 1950-2012 in order to make predictions from 2013-2015.  After making a set of initial predictions based off of the closing price from prior days, I ran additional sets of predictions which take into account day of the week and unemployment rate.

## Preparing the data

```python
import pandas as pd

data = pd.read_csv('sphist.csv')
print(data.dtypes)
data.head()
```

First I will convert the 'Date' column to a datetime data type and reorganize the data so its in ascending order by date

```python
from datetime import datetime

#Update Date column to datetime
data['Date'] = pd.to_datetime(data['Date'])

#Reorder data so the oldest date comes first
data = data.sort_values(by='Date')
```

## Creating indicators from prior closing prices

Due to the timeseries nature of this dataset, each day's closing price is somewhat dependant upon the closing price from previous days. Below, I will define a few functions that will create indicators that can make the model more accurate.  One of these functions will compute the average closing price of the previous x-number of days. A second function will compute the standard deviation in closing price from the previous x-number of days.

```python
def compute_avg_x_days(n_days, df, feature):
    n_day_avg = []
    for index in range(len(df)):
        if index < n_days:
            n_day_avg.append(0)
        else:
            past_x_days = df.iloc[(index - n_days) : index]
            avg_close = past_x_days[feature].mean()
            n_day_avg.append(avg_close)
    new_column = 'avg_' + feature + '_' + str(n_days) + '_days'
    df[new_column] = pd.Series(n_day_avg).values
    
def std_x_days(n_days, df, feature):
    n_day_avg = []
    for index in range(len(df)):
        if index < n_days:
            n_day_avg.append(0)
        else:
            past_x_days = df.iloc[(index - n_days) : index]
            avg_close = past_x_days[feature].std()
            n_day_avg.append(avg_close)
    new_column = 'stdev_' + feature + '_' + str(n_days) + '_days'
    df[new_column] = pd.Series(n_day_avg).values
```

```python
compute_avg_x_days(5, data, 'Close')
compute_avg_x_days(365, data, 'Close')
std_x_days(5, data, 'Close')
std_x_days(365, data, 'Close')
data.head(10)
```

Now that I have created a few indicators based off of historical data, there will be some rows containing null values where there wasn't yet enough data to calculate these indicators.  I will remove these rows.

```python
data_non_nulls = data[(data != 0).all(1)]
data_non_nulls.head()
```

## Making initial predictions

In order to make predictions on the updated dataset, I will first split the data into a training set and a test set. The training set will contain all dates prior to 2013 and the test set will contain all dates starting in 2013.  I will use a linear regression model in order to make predictions and I will use Mean Absolute Error as my error metric. I will be using Mean Absolute Error instead of Mean Sqaured Error because it will give a more accurate indication as to how far off I am from the true closing price.

```python
#Make DF for Train (Before 2013) and Test (After 2013)
train = data_non_nulls[data_non_nulls['Date'] < '2013-01-01']
test = data_non_nulls[data_non_nulls['Date'] >= '2013-01-01']

#Confirm we're not missing any rows
print('test rows + train rows = ' + str(train.shape[0] + test.shape[0]))
print('full dataset rows = ' + str(data_non_nulls.shape[0]))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

#Drop all of our original columns with information related to the closing price
features = [i for i in data_non_nulls.columns if i not in ('Close', 'High', 'Low', 'Open', 'Volume',
                                             'Adj Close', 'Date')]
#Run LinearRegression() model to predict 'Close' column of test dataframe
lr = LinearRegression()
lr.fit(train[features], train['Close'])
predictions = lr.predict(test[features])
test['predictions'] = predictions
test.head(5)
```

```python
#Calculate Mean Absolute Error
mae = mean_absolute_error(test['Close'], test['predictions'])
print('MAE: ' + str(mae))
```

Initially we have a Mean Absolute Error of 16.13 with a corresponding graph below which plots both the actual and predicted closing prices.

```python
#Plot closing prices in comparison to prediction
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(18,7))
plt.plot(test['Date'], test['Close'], label = 'Closing Price')
plt.plot(test['Date'], test['predictions'], label = 'Predicted Closing Price')
plt.legend(prop={'size': 15})
plt.show()
```

## Factoring in Day of the Week

In an effort to reduce the value of our error metric, I will try factoring in some additional data that could have an impact on the S&P 500 closing price.  I will first check to see if day of the week will have an impact on closing price.  Some economists believe that day of the week can impact stock prices as people are more optimistic and likely to buy heading into the weekend, and can be more pessimistic and likely to sell earlier in the week. 

```python
#Create 'day_of_week' column
data_non_nulls['day_of_week'] = data_non_nulls['Date'].apply(lambda x: x.strftime('%A'))

#Create dummy variables for each day of the week 
day_of_week_df = pd.get_dummies(data_non_nulls['day_of_week'])

#Remove the initial 'day_of_week' column
data_with_day_of_week = pd.concat([data_non_nulls, day_of_week_df], axis=1).drop('day_of_week', axis=1)

data_with_day_of_week.head(5)
```

```python
#Make train (Before 2013) and test (After 2013) dataframes
train = data_with_day_of_week[data_with_day_of_week['Date'] < '2013-01-01']
test = data_with_day_of_week[data_with_day_of_week['Date'] >= '2013-01-01']
```

```python
#Drop all of our original columns with information related to the closing price
features = [i for i in data_with_day_of_week.columns if i not in ('Close', 'High', 'Low', 'Open', 'Volume',
                                             'Adj Close', 'Date')]

#Run LinearRegression() model to predict 'Close' column of test dataframe
lr = LinearRegression()
lr.fit(train[features], train['Close'])
predictions = lr.predict(test[features])
test['predictions'] = predictions

#Calculate Mean Absolute Error
mae = mean_absolute_error(test['Close'], test['predictions'])
print('MAE: ' + str(mae))
```

By adding in day of the week, our Mean Absolute Error increased. However, this has made us aware of an issue with the current analysis. Instead of predeicting the closing price when including factors like day of the week, we should be creating a new target like percentage change from the previous day. Based off of the precentage change from the previous day we should then predict a closing price. Simply adding day of week into our calculations isn't a good indicator because if the day happens to be a Friday as an example, closing prices from all prior Fridays will skew our results. It would be a much better indicator to see how much of an impact Friday has on the overall percentage change rather than use it to predict our overall closing price. 

## Using Day of Week to predict percentage change in closing price

```python
#Function to calculate the percentage change of closing price on a given day
def calc_close_percent_change(data):
    percent_changes = []
    for index in range(len(data)):
        if index > 0:
            prior_row = data.iloc[index - 1]
            current_row = data.iloc[index]
            percent_changes.append((current_row['Close'] - prior_row['Close']) / prior_row['Close'])
        else:
            percent_changes.append(0)
    return pd.Series(percent_changes).values

#Calculate percentage change of closing price for all rows in our dataset
price_changes = calc_close_percent_change(data)

#Add percentage change in closing price column to our dataset
data['close_price_change'] = price_changes

#Confirm that our new column is accurate
data[['Close', 'close_price_change']].head()
```

```python
#Add column to compute the average change in closing price over the previous 5 days
compute_avg_x_days(5, data, 'close_price_change')

#Add column to compute the standard deviation in closing price over the previous 5 days
std_x_days(5, data, 'close_price_change')
```

```python
#drop nulls
data_non_nulls = data[(data != 0).all(1)]

#Add dummy columns for day of the week
data_non_nulls['day_of_week'] = data_non_nulls['Date'].apply(lambda x: x.strftime('%A'))
day_of_week_df = pd.get_dummies(data_non_nulls['day_of_week'])
data_with_day_of_week = pd.concat([data_non_nulls, day_of_week_df], axis=1).drop('day_of_week', axis=1)
```

```python
#Make train (Before 2013) and test (After 2013) dataframes
train = data_with_day_of_week[data_with_day_of_week['Date'] < '2013-01-01']
test = data_with_day_of_week[data_with_day_of_week['Date'] >= '2013-01-01']

#Drop all of our original columns with information related to the closing price
features = [i for i in data_with_day_of_week.columns if i not in ('Close', 'High', 'Low', 'Open', 'Volume', 'Adj Close', 'Date', 'close_price_change')]
lr = LinearRegression()
lr.fit(train[features], train['close_price_change'])

#Run LinearRegression() model to predict 'close_price_change' column of test dataframe
predictions = lr.predict(test[features])

#Add predictions to test dataframe
test['predict_price_change'] = predictions
```

```python
#Define function to predict closing price based off of the predicted change in closing price
def predict_close(data):
    predictions = []
    for index in range(len(data)):
        if index > 0:
            prior_row = data.iloc[index - 1]
            current_row = data.iloc[index]
            predictions.append(prior_row['Close'] * (1 + current_row['predict_price_change']))
        else:
            predictions.append(0)
    return pd.Series(predictions).values

```

## Factoring in Unemployment

In an effort to further improve our model, I will include data on the US unemployment rate to see if this has an impact on stock prices.  The unemployment rate is often used as an indicator of overall economic health with the idea being that lower unemployment leads to a healthy economy where people can afford to buy more stocks. The data I have included comes from the [US Bureau of Labor Statistics](https://www.bls.gov/).

```python
unemployment = pd.read_csv('Unemployment_Data.csv')
print(unemployment.dtypes)
unemployment.head()
```

Before adding this data to our S&P 500 dataset, I will clean it up by converting the numeric columns to percentage values, removing null values, and creating a Date colum which accounts for both month and year.

```python
#Convert numeric columns to percentages
for column in ['Unemployment Rate', '1-Month Percent Change', '3-Month Change', '12-Month Change']:
    unemployment[column] = unemployment[column].apply(lambda x: x / 100)

#Convert Period column to a number representing the corresponding month
unemployment['Period'] = unemployment['Period'].str.replace('M', '').astype(int)
#Create a date column that includes month and year
unemployment['Date'] = unemployment['Period'].map(str) + ' - ' + unemployment['Year'].map(str)

#Rename columns so they're easier to work with and drop the individual 'Year' and 'Period'(month) columns 
renamed_columns = {
    'Unemployment Rate': 'unemployment',
    '1-Month Percent Change': 'change_1_mo',
    '3-Month Change': 'change_3_mo',
    '12-Month Change': 'change_12_mo'
}
unemployment = unemployment.drop(columns=['Year', 'Period']).rename(columns=renamed_columns)

#Check for null values
print(unemployment.isnull().sum())
```

```python
unemployment.head()
```

I will now combine the unemployment data with the main S&P500 dataset. I will need to do some additional feature engineering so that each date in the S&P 500 dataset includes unemployment data from the previous month.  Unemployment statistics are not calculated until the end of the month so including the current month's data would mean that I've included information from the future which would cause our results to be biased. Another reason why I want to link each date with the prior month's data is becuase this information affects overall investor confidence in the economy which in turn impacts the S&P 500 price. For example, if investors learn on March 1st that unemployment went up by 10% in the month of February, then they will have less confidence in the economy for the entire month of March which should cause S&P 500 prices to drop over the course of the month.

```python
#Create separate year and month columns for S&P 500 data
data_with_day_of_week['year'] = pd.DatetimeIndex(data_with_day_of_week['Date']).year
data_with_day_of_week['month'] = pd.DatetimeIndex(data_with_day_of_week['Date']).month
data_with_day_of_week.head(10)

#Define function to identify the prior month for each row on which we will join unemployment data
def calc_prior_month(row):
    if row['month'] == 1:
        year = str(row['year'] - 1)
        month = '12'
    else:
        year = str(row['year'])
        month = str(row['month'] - 1)
    return month + ' - ' + year

#Add column indicating the previous month ('unemployment_id') to each row
data_with_day_of_week['unemployment_id'] = data_with_day_of_week.apply(lambda x: calc_prior_month(x), axis=1)
#Drop individual yeara and month columns from S&P 500 data
data_with_day_of_week = data_with_day_of_week.drop(columns=['year', 'month'])
data_with_day_of_week.head(5)
```

```python
#Rename Date column in the unemployment dataframe so we can join it with the S&P 500 data
unemployment = unemployment.rename(columns={'Date': 'unemployment_id'})

#Join S&P 500 data with unemployment data on 'unemployment_id' column
unemploymenet_combined = pd.merge(data_with_day_of_week, unemployment, on='unemployment_id', how='left')
#Drop 'unemployment_id' after dataframes have been joined
unemploymenet_combined = unemploymenet_combined.drop(columns=['unemployment_id'])
```

## Predictions with Unemployment Data

I will now factor in the unemployment to predict the percentage change in closing price and use these results to predict daily closing prices. 

```python
#Make train (Before 2013) and test (After 2013) dataframes
train = unemploymenet_combined[unemploymenet_combined['Date'] < '2013-01-01']
test = unemploymenet_combined[unemploymenet_combined['Date'] >= '2013-01-01']

#Drop all of our original columns with information related to the closing price
features = [i for i in unemploymenet_combined.columns if i not in 
            ('Close', 'High', 'Low', 'Open', 'Volume', 'Adj Close', 'Date', 'close_price_change', 'change_12_mo')]
lr = LinearRegression()
lr.fit(train[features], train['close_price_change'])

#Run LinearRegression() model to predict 'close_price_change' column of test dataframe
predictions = lr.predict(test[features])
#Add predictions to test dataframe
test['predict_price_change'] = predictions

#Create a new dataframe that only uses relevant information for predicting closing price
predicted_close_df = test[['Date', 'Close', 'close_price_change', 'predict_price_change']]

#Generate predictions using predict_close() function from earlier and add them to our predicted_close_df
predictions = predict_close(predicted_close_df)
predicted_close_df['predict_close'] = predictions

#remove first row as it contains a null value
predicted_close_df = predicted_close_df[1:]
predicted_close_df.head()
```

```python
#Calculate Mean Absolute Error
mae = mean_absolute_error(predicted_close_df['Close'], predicted_close_df['predict_close'])
print('MAE: ' + str(mae))

plt.figure(figsize=(18,7))
plt.plot(predicted_close_df['Date'], predicted_close_df['Close'], label = 'Closing Price')
plt.plot(predicted_close_df['Date'], predicted_close_df['predict_close'], label = 'Predicted Closing Price')
plt.legend(prop={'size': 15})
plt.show()
```
